该题目是 LeetCode 第 4 题 **“寻找两个正序数组的中位数”**，要求在两个已经排好序的数组中找到它们合并后的中位数。

---

### 题目描述：
给定两个大小分别为 

m

 和 

n

 的正序（从小到大）数组 

nums1

 和 

nums2

。请你找出并返回这两个正序数组的 **中位数**。

#### 示例：
1. 输入：

nums1 = [1, 3]

, 

nums2 = [2]

  
   输出：`2.0`  
   解释：合并后的数组为 `[1, 2, 3]`，中位数是 `2`。

2. 输入：

nums1 = [1, 2]

, 

nums2 = [3, 4]

  
   输出：`2.5`  
   解释：合并后的数组为 `[1, 2, 3, 4]`，中位数是 `(2 + 3) / 2 = 2.5`。

3. 输入：

nums1 = [0, 0]

, 

nums2 = [0, 0]

  
   输出：`0.0`

4. 输入：

nums1 = []

, 

nums2 = [1]

  
   输出：`1.0`

5. 输入：

nums1 = [2]

, 

nums2 = []

  
   输出：`2.0`

---

### 解题思路：
这道题要求时间复杂度为 **O(log(min(m, n)))**，因此不能直接合并两个数组后排序，而是需要使用 **二分查找**。

#### 核心思路：
1. **中位数的定义**：
   - 如果两个数组的总长度是奇数，中位数是合并后数组的中间值。
   - 如果总长度是偶数，中位数是合并后数组的中间两个值的平均值。

2. **二分查找**：
   - 假设我们在较短的数组 

nums1

 上进行二分查找，找到一个分割点 

i

，使得：
     - 左边部分的最大值 ≤ 右边部分的最小值。
   - 分割点 

i

 和 

j

 满足：

i + j = (m + n + 1) / 2

。
   - 通过调整 

i

 的位置，找到满足条件的分割点。

3. **边界条件**：
   - 如果 

i

 太小，说明左边的值太小，需要增大 

i

。
   - 如果 

i

 太大，说明左边的值太大，需要减小 

i

。

---

### 时间复杂度：
- **O(log(min(m, n)))**：二分查找的复杂度取决于较短数组的长度。

### 空间复杂度：
- **O(1)**：只使用了常数级别的额外空间。

这是一个经典的二分查找问题，适合用来练习高效算法设计。

```java
class Solution {
    public double findMedianSortedArrays(int[] nums1, int[] nums2) {
    int m = nums1.length;
    int n = nums2.length;

    // 确保 nums1 是较短的数组
    if (m > n) {
        int[] temp = nums1;
        nums1 = nums2;
        nums2 = temp;
        int tmp = m;
        m = n;
        n = tmp;
    }

    int iMin = 0, iMax = m, halfLen = (m + n + 1) / 2;
    while (iMin <= iMax) {
        int i = (iMin + iMax) / 2;
        int j = halfLen - i;

        // 调整分割点
        if (i < iMax && nums2[j - 1] > nums1[i]) {
            iMin = i + 1; // i 太小
        } else if (i > iMin && nums1[i - 1] > nums2[j]) {
            iMax = i - 1; // i 太大
        } else {
            // 找到合适的分割点
            int maxLeft;
            if (i == 0) {
                maxLeft = nums2[j - 1];
            } else if (j == 0) {
                maxLeft = nums1[i - 1];
            } else {
                maxLeft = Math.max(nums1[i - 1], nums2[j - 1]);
            }

            if ((m + n) % 2 == 1) {
                return maxLeft; // 奇数时直接返回
            }

            int minRight;
            if (i == m) {
                minRight = nums2[j];
            } else if (j == n) {
                minRight = nums1[i];
            } else {
                minRight = Math.min(nums1[i], nums2[j]);
            }

            return (maxLeft + minRight) / 2.0; // 偶数时返回平均值
        }
    }
    return 0.0;
}
}
Solution solution = new Solution();
int[] nums1 = {1, 3};
int[] nums2 = {2};
double result = solution.findMedianSortedArrays(nums1, nums2);
System.out.println(result);
```
