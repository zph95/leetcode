给定整数 n ，返回 所有小于非负整数 n 的质数的数量 。

 

示例 1：

输入：n = 10
输出：4
解释：小于 10 的质数一共有 4 个, 它们是 2, 3, 5, 7 。
示例 2：

输入：n = 0
输出：0
示例 3：

输入：n = 1
输出：0
 

提示：

0 <= n <= 5 * 106

当前的算法使用了暴力方法来判断每个数字是否为质数，时间复杂度较高，约为 \(O(n \sqrt{n})\)。可以通过 **埃拉托色尼筛法（Sieve of Eratosthenes）** 来优化，时间复杂度可以降低到 \(O(n \log \log n)\)。

以下是优化后的代码：

```java
/*
 * @lc app=leetcode.cn id=204 lang=java
 *
 * [204] 计数质数
 */

// @lc code=start
class Solution {
    public int countPrimes(int n) {
        if (n <= 2) return 0;

        boolean[] isPrime = new boolean[n];
        for (int i = 2; i < n; i++) {
            isPrime[i] = true;
        }

        for (int i = 2; i * i < n; i++) {
            if (isPrime[i]) {
                for (int j = i * i; j < n; j += i) {
                    isPrime[j] = false;
                }
            }
        }

        int count = 0;
        for (int i = 2; i < n; i++) {
            if (isPrime[i]) count++;
        }

        return count;
    }
}
// @lc code=end
```

### 优化点说明：
1. **埃拉托色尼筛法**：
   - 创建一个布尔数组 `isPrime`，初始时假设所有数都是质数。
   - 从 2 开始，将每个质数的倍数标记为非质数。
   - 只需筛选到 \(\sqrt{n}\) 即可，因为更大的倍数已经被更小的质数筛选过。

2. **时间复杂度**：
   - 外层循环最多运行到 \(\sqrt{n}\)。
   - 内层循环标记倍数的复杂度为 \(O(n / 2 + n / 3 + n / 5 + \dots)\)，总和为 \(O(n \log \log n)\)。

3. **空间复杂度**：
   - 需要一个大小为 \(n\) 的布尔数组，空间复杂度为 \(O(n)\)。

这个优化方法在处理大范围的质数计数时非常高效。
