 * 条路线。因此，要想在两座不同城市之间旅行只有唯一一条路线可供选择（路线网形成一颗树）。去年，交通运输部决定重新规划路线，以改变交通拥堵的状况。

```java
import java.util.ArrayDeque;
import java.util.ArrayList;
import java.util.HashMap;
import java.util.List;
import java.util.Map;
import java.util.Queue;
import java.util.Set;
import java.util.HashSet;

class Solution {

    Set<Integer> attachCities = new HashSet<>();
    int count = 0; int[][] roads;

    public int minReorder(int n, int[][] connections) {
        roads = new int[n][n];
        for (int i = 0; i < n - 1; i++) {
            roads[connections[i][0]][connections[i][1]] = 1;
        }
        Queue<Integer> cityQueue = new ArrayDeque<>();
        attachCities.add(0);
        List<Integer> linkCities = getLinkCity( n, 0);
        cityQueue.addAll(linkCities);
        while (!cityQueue.isEmpty()) {
            Integer city = cityQueue.poll();
            List<Integer> tmp = getLinkCity( n, city);
            cityQueue.addAll(tmp);
        }
        return count;
    }

    List<Integer> getLinkCity( int n, int i) {
        List<Integer> result = new ArrayList<>();
        for (int j = 0; j < n; j++) {
            if (roads[j][i] == 1 && !attachCities.contains(j)) {
                result.add(j);
                attachCities.add(j);
            }
            else if (roads[i][j] == 1 && !attachCities.contains(j)) {
                result.add(j);
                count++;
                attachCities.add(j);
                roads[i][j] = 0;
                roads[j][i] = 1;
            }
        }
        return result;
    }
}
        int n =2 ;
        int[][] connections=new int[1][2];
        connections[0][0]=0;
        connections[0][1]=1;
        Solution solution=new Solution();
        System.out.println(solution.minReorder(n, connections));
```
