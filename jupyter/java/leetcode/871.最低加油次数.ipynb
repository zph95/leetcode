这道题目是 LeetCode 第 871 题 **"最低加油次数"**，其描述如下：

### 题目翻译
一辆汽车从起始位置出发，目标是到达距离起始位置 `target` 英里的目的地。

沿途有若干个加油站，使用数组 `stations` 表示，其中 `stations[i] = [positioni, fueli]` 表示第 `i` 个加油站位于距离起始位置 `positioni` 英里处，并且该加油站可以提供 `fueli` 升的汽油。

汽车的油箱容量是无限的，初始时有 `startFuel` 升的汽油。汽车每行驶 1 英里会消耗 1 升汽油。当汽车到达一个加油站时，可以选择停下来加油，将加油站的所有汽油加到车里。

返回汽车到达目的地所需的最少加油次数。如果无法到达目的地，返回 `-1`。

---

### 示例
#### 示例 1:
```plaintext
输入: target = 100, startFuel = 10, stations = [[10,60],[20,30],[30,30],[60,40]]
输出: 2
解释: 
我们可以在以下位置加油：
- 在位置 10 加 60 升油，总油量为 70。
- 在位置 60 加 40 升油，总油量为 110。
最终到达目的地。
```

#### 示例 2:
```plaintext
输入: target = 1, startFuel = 1, stations = []
输出: 0
解释: 不需要加油即可到达目的地。
```

#### 示例 3:
```plaintext
输入: target = 100, startFuel = 1, stations = [[10,100]]
输出: -1
解释: 无法到达目的地。
```

---

### 题目要求
1. **目标**: 找到到达目的地所需的最少加油次数。
2. **返回值**: 如果无法到达目的地，返回 `-1`。

---

### 解题思路
1. **动态规划**:
   - 定义 `dp[i]` 表示加油 `i` 次后，汽车能到达的最远距离。
   - 初始时，`dp[0] = startFuel`。
   - 遍历每个加油站，更新 `dp` 数组，计算加油后能到达的最远距离。
   - 如果某次加油后能到达目标距离 `target`，返回加油次数。

2. **贪心算法**:
   - 使用一个最大堆（优先队列）存储可以加的油量。
   - 每次到达一个加油站时，将油量加入堆中。
   - 如果油量不足以到达下一个加油站或目的地，从堆中取出最大油量加油，直到可以继续前进或无法到达。

动态规划的时间复杂度为 `O(n^2)`，贪心算法的时间复杂度为 `O(n log n)`，其中 `n` 是加油站的数量。

````python
/*
 * @lc app=leetcode.cn id=871 lang=java
 *
 * [871] 最低加油次数
 */

// @lc code=start
class Solution {
    public int minRefuelStops(int target, int startFuel, int[][] stations) {
        int n = stations.length;
        long[] dp = new long[n + 1];

        dp[0] = startFuel;
        for (int i = 0; i < n; i++) {
            for (int t = i; t >= 0; t--) {
                if (dp[t] >= stations[i][0]) {
                    dp[t + 1] = Math.max(dp[t + 1], dp[t] + stations[i][1]);
                }
            }
        }
       
        for (int i = 0; i <= n; i++) {
            if (dp[i] >= target) {
                return i;
            }
        }
        return -1;
    }
}
// @lc code=end

````
